
# Relatório das Respostas — Pesquisa de Intenção
## O Lugar que Ainda Não Existe
**Total de respostas:** 19

---

## 1. Perfil dos Respondentes
- Origem: Maioria de Curitiba (bairros variados)
- Público: Pessoas ligadas a projetos criativos, tecnologia, cultura maker, autônomos ou que frequentam coworkings/home office.

---

## 2. Interesse em Frequentar
100% dos participantes demonstraram interesse; 79% com entusiasmo total.

![Gráfico Interesse](grafico_interesse.png)

---

## 3. Frequência Imaginada de Uso
A maioria pretende usar o espaço semanalmente ou mais.

![Gráfico Frequência](grafico_frequencia.png)

---

## 4. Tipo de Uso Mais Desejado
O espaço é visto tanto para trabalho individual quanto para encontros, eventos e oficinas.

![Gráfico Tipo de Uso](grafico_tipo_uso.png)

---

## 5. Projetos Que Gostariam de Realizar
- Tecnologia, cultura maker, prototipagem, impressão 3D
- Arte, música, oficinas, eventos culturais
- Ações sociais, projetos colaborativos
- Flash tattoos, workshops, clubes de leitura

---

## 6. Recursos e Ambientes Desejados
- Café/bar/cafeteira (unanimidade)
- Sala de reunião
- Estúdio de gravação/som/podcast
- Oficinas criativas/ambiente maker
- Impressora 3D/FabLab
- Espaço externo/verde
- Biblioteca, espaço esportivo

---

## 7. O Que Tornaria o Espaço Irresistível
- Café de verdade, bom ambiente e colegas
- Senso de pertencimento
- Flexibilidade e acolhimento
- Espaço verde e relaxante
- Estímulo à colaboração e fácil acesso a recursos

---

## 8. Faixa de Preço Aceitável por Mês
58% pagariam até R$ 200/mês; há nicho premium.

![Gráfico Faixa de Preço](grafico_preco.png)

---

## 9. Interesse em Planos Longos/Desconto
53% já pensam em planos anuais, 37% negociariam desconto por fidelidade.

![Gráfico Planos Longos](grafico_planos.png)

---

## 10. Interesse em Colaborar na Construção
89% querem colaborar ativamente; há interesse em parceria, voluntariado e até investimento.

![Gráfico Colaboração](grafico_colaborar.png)

---

## 11. Sugestões, Ideias e Comentários
- Interesse genuíno em colaborar
- Desejo de FabLab, área para crianças, estrutura para arte/música
- Espaço pet friendly como diferencial
- Comunidade acima de competição

---

## Principais Insights
1. Demanda real e engajada: necessidade de convivência criativa e colaboração presencial.
2. Café + Maker + Eventos: tríade essencial.
3. Diversidade de perfis e usos.
4. Modelo de assinatura aceito, com preço médio até R$ 200/mês.
5. Colaboração espontânea: base para programa de fundadores/embaixadores.

---

## Recomendações
- MVP: café/bar, sala de reunião, estúdio/podcast, espaço maker, eventos e comunidade.
- Testar planos mensais, anuais, passes diários, fundador vitalício.
- Envolver interessados desde já: grupos de cocriação, rodadas de conversa, voluntariado.
- Comunicação: destacar pertencimento, mostrar projetos reais.
- Diferenciais: pet friendly, áreas externas, biblioteca colaborativa, agenda cultural.

---

## Resumo Final
A comunidade está pronta para apoiar e co-criar o projeto. O Lugar que Ainda Não Existe pode nascer com uma base forte e colaborativa. Próximos passos: envolver a rede, testar eventos-piloto, ajustar o modelo de preço e continuar ouvindo o público.
